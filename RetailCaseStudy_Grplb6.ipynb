```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
#from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
def macro_data():
    macro = pd.read_csv('MacroEconomicData.csv')
    lst = macro['Year-Month'].values
    year = []
    month = []
    mon = 0
    for value in lst:
        splits = value.split('-')
        year.append(int(splits[0]))
        if splits[1] == ' Jan':
            mon = 1
        if splits[1] == ' Feb':
            mon = 2
        if splits[1] == ' Mar':
            mon = 3
        if splits[1] == ' Apr':
            mon = 4
        if splits[1] == ' May':
            mon = 5
        if splits[1] == ' Jun':
            mon = 6
        if splits[1] == ' Jul':
            mon = 7
        if splits[1] == ' Aug':
            mon = 8
        if splits[1] == ' Sep':
            mon = 9
        if splits[1] == ' Oct':
            mon = 10
        if splits[1] == ' Nov':
            mon = 11
        if splits[1] == ' Dec':
            mon = 12
    
        month.append(mon)
    macro['Year'] = year
    macro['Month'] = month
    macro_data = macro.drop(['Year-Month','PartyInPower'],axis = 1)
    for i,val in enumerate(macro_data['AdvertisingExpenses (in Thousand Dollars)']):
        if val == '?':
            macro_data['AdvertisingExpenses (in Thousand Dollars)'][i] = 0
    
    
    return macro_data
```

```python
macro_data = macro_data()
```

```python
for i,val in enumerate(data['Sales(In ThousandDollars)']):
    if val=='NaN':
        if data['ProductCategory'][i] == 'WomenClothing':
            data['Sales(In ThousandDollars)'][i]=mean_women
        if data['ProductCategory'][i] == 'MenClothing':
            data['Sales(In ThousandDollars)'][i]=mean_men
        if data['ProductCategory'][i] == 'OtherClothing':
            data['Sales(In ThousandDollars)'][i]=mean_others
```
